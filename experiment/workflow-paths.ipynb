**Stage 1: Basic Chatbot (LLM only)**

**Objective:** Chatbot that answers queries using only the LLM.

**Required Nodes:**

1. **User Input Node** → captures the user query.
2. **LLM Node** → generates a response based on the input.
3. **Output Node** → sends the answer back to the user.

**Flow:**
User Input → LLM Node → Output



**Stage 2: RAG-based Chatbot**

**Objective:** Use retrieval-augmented generation for knowledge-based answers.

**Required Nodes:**

1. **User Input Node**
2. **RAG Retriever Node** → fetch relevant context from a vector database.
3. **LLM Node** → uses the retrieved context + user query to generate an answer.
4. **Output Node**

**Flow:**
User Input → RAG Retriever → LLM Node → Output



**Stage 3: Separate MCP Tool Agent**

**Objective:** Add external tool capabilities for specialized actions.

**Required Nodes:**

1. **User Input Node**
2. **Action Selector Node** → decides between LLM answer or MCP tool.
3. **MCP Tool Node** → performs specific action (calculator, API, database).
4. **Observation Node** → captures tool output.
5. **LLM Node** → optionally processes tool output for final answer.
6. **Output Node**

**Flow:**
User Input → Action Selector → [LLM Node OR MCP Tool Node → Observation → LLM Node] → Output



**Stage 4: Integrated Multi-Agent Chatbot**

**Objective:** Combine LLM, RAG, and MCP agents into one intelligent system.

**Required Nodes:**

1. **User Input Node**
2. **Action Selector Node** → decides among LLM, RAG, or MCP.
3. **RAG Retriever Node** → context retrieval.
4. **MCP Tool Node** → external action.
5. **Observation Node** → collects results from RAG or MCP.
6. **LLM Node** → generates final response considering all observations.
7. **Output Node**
8. **Logging/Observability Node** → tracks actions, results, and errors.

**Flow:**
<div style="background-color: #000000; color: #ffffff; padding: 15px; border-radius: 8px; border: 2px solid #333333; margin: 10px 0; font-family: monospace; white-space: pre;">

User Input
      ↓
Action Selector → [LLM Node OR RAG Retriever → Observation → LLM Node OR MCP Tool → Observation → LLM Node]
      ↓
Output
      ↓
Logging/Observability
